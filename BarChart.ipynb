```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df1 = pd.read_csv('RegionWiseExport(Goods)ForTheMonthofJuly-June2018-19.csv')

df1.head()
```

```python
df1 = df1.rename(columns={'Value US$':'Value'})

df1.head()
```

```python
a1 = df1[df1.Region.notnull()]
#a1 = a1.set_index('Region')
a1.drop('Country',axis=1,inplace=True) # Drop first column
a1.sort_values('Value', ascending = False, inplace =True)
```

```python
a1
```

```python
fig = plt.figure(figsize = (20,5))

ax1 = fig.add_subplot(1,2,1)

ax1.bar(a1.Region, a1.Value, color = 'red')
ax1.set_xticklabels(a1.Region, rotation = 60, horizontalalignment = 'right', fontsize = '12')

ax1.set_title('Amount of value Exported in different Regions ($)', fontsize = '17')
ax1.set_ylabel('Value ($)')

plt.show()
```
